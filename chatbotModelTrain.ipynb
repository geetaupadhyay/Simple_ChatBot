```python
import json
import numpy as np
import random
from nltk.stem import WordNetLemmatizer, PorterStemmer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD
from nltk.corpus import stopwords
```

```python
#cleaning the data for better use
```

```python
lemmatizer=WordNetLemmatizer()
intents=json.loads(open('intents.json').read())
words=[]
Labels=[]
documents=[]
unnecessary_letter=['?','!', '.',',']
for intent in intents["intent"]:
    for pattern in intent["patterns"]:
        word_of_list=nltk.word_tokenize(pattern)
        words.extend(word_of_list)
        documents.append((word_of_list, intent["tag"]))
        if intent["tag"] not in Labels:
            Labels.append(intent["tag"])
# print(documents, end="\n")
print(words, end="\n")
```

```python
words=[lemmatizer.lemmatize(word) for word in words if word not in unnecessary_letter]
words=sorted(set(words))
print(words)
```

```python
# stopwords.words("english")-->do not use this, data will be ambiguous
```

```python
Labels=sorted(set(Labels))
Labels
```

```python
pickle.dump(words,open("Words_set.pkl","wb"))
pickle.dump(Labels,open("Labels_set.pkl","wb"))
```

```python
#perparing data for model using BOW
```

```python
training=[]
output=[0]*len(Labels)
```

```python
for document in documents:
    BOW=[]
    word_pattern=document[0]
    word_pattern=[lemmatizer.lemmatize(word.lower()) for word in word_pattern]
    for word in words:
        if word in word_pattern:
            BOW.append(1)
        else:
            BOW.append(0)
    output_row=list(output)
    output_row[Labels.index(document[1])]=1
    training.append([BOW,output_row])
```

```python
print(training[0])
```

```python
random.shuffle(training)
training=np.array(training)
```

```python
train_x=list(training[:,0])
train_y=list(training[:,1])
#tried to move array into sections without converting it into
#numpy 2d array so it gave error of list..
#..indices must not be tuple.
```

```python
#define the model to be used
```

```python
model=Sequential()
model.add(Dense(128,input_shape=(len(train_x[0]),),activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(64,activation="relu"))
model.add(Dense(len(train_y[0]), activation="softmax"))
Gradient=SGD(lr=0.001, momentum=0.9, nesterov=True)
model.compile(loss="categorical_crossentropy", optimizer=Gradient, metrics=["accuracy"])
```

```python
model.summary()
```

```python
Model2=model.fit(np.array(train_x),np.array(train_y), epochs=500, batch_size=5, verbose=1)

```

```python
if os.path.isfile("CHATBOT_MODELS/chatbotModel2.h5") is False:
    model.save("CHATBOT_MODELS/chatbotModel2.h5", Model2)
else:
    print("already Saved!")
#CHATBOT_MODELS
```
